```python
import pandas as pd
import plotly.offline as offline
import plotly
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
from bokeh.palettes import Spectral11
from bokeh.plotting import figure, show, output_file
import plotly.offline as offline
import plotly.offline as offline
from plotly.offline import init_notebook_mode, iplot
from IPython.display import display, HTML
init_notebook_mode(connected=True)
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
df = pd.read_csv('cc_institution_details.csv',encoding = 'latin1')
```

```python
inst = list()
for state in df['state'].unique():
    inst.append(go.Bar(x = df[df.state == state]['chronname'].values.tolist(),
                       y = df[df.state == state]['grad_100_value'].values.tolist(),
                       name = state,
                       visible = False))

button = list()
for row, state in df.state.drop_duplicates().reset_index()['state'].iteritems():
    condition = [False]*len(df.state.drop_duplicates())
    condition[row] = True
    button.append(dict(label = state,method = 'update',args = [{'visible': condition},{'title': 'Graduation rate for institutes in ' + state}]))

updatemenus=list([
    dict(
        buttons = button,
        pad = {'l': 0, 't': 0},
        #x = 0.1,
        #xanchor = 'left',
        #y = 1.0,
        yanchor = 'top',
        bgcolor = 'AAAAAA',
        active = 99,
        font = dict(size=10, color='#000000'))])

layout = dict(title='Graduation rate for institutes in different states', showlegend=False, updatemenus=updatemenus)

figure = dict(data=inst, layout=layout)
#plotly.offline.plot(figure)
iplot(figure)
```
