```python
df = data.groupby(['state'])['grad_100_value'].mean().reset_index()
```

```python

#df.state.unique()
#df[df['state']=='District of Columbia']['state']='South Carolina'
df['state'] = ['South Carolina' if x == 'District of Columbia' else x for x in df['state']]
df['code'] = [state[x] for x in df['state']]
df['code'].unique()
```

```python
for col in df.columns:
    df[col] = df[col].astype(str)
```

```python


scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
df['text'] = df['state'] + '<br>' +'Graduation Rate '+df['grad_100_value']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['code'],
        z = df['grad_100_value'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Percentage")
        ) ]

layout = dict(
        title = 'Graduation rates by State<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
#py.iplot( fig, filename='d3-cloropleth-map' )
#plotly.offline.plot(fig, filename='ch.html')
iplot(fig)
```
